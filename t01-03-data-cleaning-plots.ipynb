## HISTOGRAMA DE FRECUENCIAS

```
# Según la Regla de Sturges, el tamaño recomendado de los bins en un histograma es el siguiente
# El tamaño de la muestra es 3333
# para eso es necesario importar numpy
# 1+log2(muestra)
```

```
k = int(np.ceil(1 + np.log2(3333)))
#plt.hist(data['Day Calls'], bins = 20) # bins es para indicar el número de rangos
plt.hist(data['Day Calls'], bins = k) # bins es para indicar el número de rangos
plt.xlabel('Número de llamadas al día')
plt.ylabel('Frecuencia')
plt.title('Histograma de número de llamadas al día')
```

# BOXPLOT (DIAGRAMA DE CAJA Y BIGOTES)

```
plt.boxplot(data['Day Calls'])
plt.ylabel('Número de llamadas diarias')
plt.title('BoxPlot de las llamadas diarias')
```

```
estadisticos_basicos = data['Day Calls'].describe()
rango_intercuantilico = estadisticos_basicos['75%'] - estadisticos_basicos['25%']
rango_intercuantilico #es la altura de la caja

#también puede accederse a los cuantiles con data['Day Calls'].quantile(0.75) para el caso del cuantil de 75%
```

```
# el bigote llega a 1.5 veces el rango intercuartílico (IQR)
bigote_superior = estadisticos_basicos['75%'] + 1.5 * rango_intercuantilico
print('El bigote superior es',bigote_superior)
bigote_inferior = estadisticos_basicos['25%'] - 1.5 * rango_intercuantilico
print('El bigote inferior es', bigote_inferior)
```

Todos los puntos por fuera de los bigotes son los 'outliers' (valores fuera de lugar)
