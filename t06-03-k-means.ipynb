```
import numpy as np
```

```
data = np.random.random(90).reshape(30,3)
data
```

```
#Elegimos dos puntos al azar para determinar que esos puntos serán los centroides
c1 = np.random.choice(range(len(data)))
c2 = np.random.choice(range(len(data)))
clust_center = np.vstack([data[c1], data[c2]])
clust_center
```

```
from scipy.cluster.vq import vq
```

```
vq(data, clust_center)
#El primer resultado es el clúster al que pertenece cada punto
#El segundo resultado nos da la distancia de cada una de las observaciones a su cluster
```

```
from scipy.cluster.vq import kmeans
```

```
kmeans(data, clust_center)
# Ojo que el resultado no son los centros de los clusters sino los baricentros del conjunto de datos que pertenecen a un 
# El segundo número es el error cuadrático medio normalizado
```

```
kmeans(data, 2) #Organizados en dos clusters
```
