A continuación se van a eliminar aquellas columnas que no se consideran interesantes para entrenar el modelo, ya que no aportan nada. Serán el nombre, el id, la maratón y la categoría. OJO que los nombres tienen que estar tal cual (case sensitive). En los datos, existía Category y CATEGORY, y el que eliminaremos es este último.

```
datos_maraton = datos_maraton.drop(columns=['Name'])
datos_maraton = datos_maraton.drop(columns=['id'])
datos_maraton = datos_maraton.drop(columns=['Marathon'])
datos_maraton = datos_maraton.drop(columns=['CATEGORY'])
datos_maraton
```

Se debería comprobar si existen datos nulos, ya que la predicción de modelos no funciona correctamente si existen datos nulos

```
datos_maraton.isna().sum()
```

Hemos visto que existen 6 datos en los que la Category, por lo que tenemos que tomar la decisión de con qué rellenar los datos que son nulos, o si eliminar directamente ese registro. Pondremos 0 en los CrossTraining y las 6 que tienen valor nulo en Category y Wall21 directamente las eliminaremos.

```
datos_maraton['CrossTraining'] = datos_maraton['CrossTraining'].fillna(0)
datos_maraton
```

```
datos_maraton = datos_maraton.dropna(how='any')
datos_maraton
```

El modelo de ML trabaja mejor si todos los datos son numéricos, así que conviene codificarlo si existe texto en alguno de los datos.

```
datos_maraton['CrossTraining'].unique()
```

```
valores_cross = {"CrossTraining": {'ciclista 1h':1, 'ciclista 3h':2, 'ciclista 4h':3, 'ciclista 5h': 4, 'ciclista 13h':5}}
datos_maraton.replace(valores_cross,inplace=True)
datos_maraton
```

```
datos_maraton['Category'].unique()
```

```
valores_cross = {"Category": {'MAM':1, 'M45':2, 'M40':3, 'M50': 4, 'M55':5, 'WAM':6}}
datos_maraton.replace(valores_cross,inplace=True)
datos_maraton
```

**Entrenamiento de modelo ML**

Ahora que ya hemos tratado los datos y vemos que son adecuados, comenzamos con el entrenamiento del modelo de ML.

Los valores que tenemos, se separarán un 80% para hacer el entrenamiento y un 20% para hacer el test.

```
datos_entrenamiento = datos_maraton.sample(frac=0.8, random_state=0)
datos_test = datos_maraton.drop(datos_entrenamiento.index)
```

```
datos_entrenamiento
```

```
datos_test
```

Separamos de los datos la variable que queremos predecir en un array independiente

```
etiquetas_entrenamiento = datos_entrenamiento.pop('MarathonTime')
etiquetas_test = datos_test.pop('MarathonTime')
```

```
datos_entrenamiento
```

```
etiquetas_entrenamiento
```

```
datos_test
```

```
etiquetas_test
```

Ahora mismo ya tenemos datos de entrenamiento y sus resultados de entrenamiento asociados, y datos de test y sus resultados esperados. Con esos 4 arrays (o matrices) ya podemos proceder al entrenamiento del modelo.
Lo haremos con una regresión lineal, que es la relación más simple.

```
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(datos_entrenamiento, etiquetas_entrenamiento)
```

```
predicciones = modelo.predict(datos_test)
predicciones
```

Como estos resultados nos tendrían que dar los etiquetas_test, podemos hacer la comparación de ambos arrays para poder ver cómo de bueno es el modelo entrenado

```
import numpy as np
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(etiquetas_test,predicciones))
print("Error porcentual: %f" % (error*100))
```

Esto lo que significa que existe un 11% de margen de error

Ahora podríamos generarnos un nuevo corredor y poder predecir con el modelo qué tiempo haría

```
nuevo_corredor = pd.DataFrame(np.array([[1,400,15,0,1.4]]),columns=['Category','km4week', 'sp4week', 'CrossTraining', 'Wall21'])
nuevo_corredor
```
