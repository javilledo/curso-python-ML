#Linear Support Vector Classifier


```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from sklearn import svm
```

```
X = [1, 5, 1.5, 8,   1,  9]
Y = [2, 8, 1.8, 8, 0.6, 11]
```

```
plt.scatter(X,Y)
```

```
data = np.array([X,Y]).T
data
```

```
target = [0, 1, 0, 1, 0, 1] #Se etiqueta cada uno de los puntos según el grupo al que pertenecen. Bajo a la izquierda es el grupo 0 y arriba a la derecha es el grupo 1
```

```
plt.scatter(X, Y, c = target)
```

```
classifier = svm.SVC(kernel = 'linear', C = 1.0)
classifier.fit(data, target)
```

```
classifier.predict([[0.58, 0.67], [10.32, 12.67]]) #Probamos la predicción para varios puntos determinados, devolviéndose un array con la categoría a la que se predice que pertenecerían
```

* Modelo w0 * x + w1 * y + intercept = 0 (Hiperplano, plano si es en 3D, recta si es para 2D)

* Ecuación del hiperplano en 2D: y = a * x + b

* Despejando de la primera se obtiene el siguiente cálculo de a y b

```
w = classifier.coef_[0]
w
```

```
a = -w[0]/w[1] #Pendiente de la recta
a
```

```
b = -classifier.intercept_[0]/w[1] #Ordenada en el origen
b
```

```
xx = np.linspace(0,10)
yy = a * xx + b
```

```
plt.plot(xx,yy, 'k-', label = 'Hiperplano de separación')
plt.scatter(X, Y, c = target)
plt.legend()
```
