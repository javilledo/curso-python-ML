```
import tensorflow as tf
```

```
x1 = tf.constant([1, 2, 3, 4, 5])
x2 = tf.constant([6, 7, 8, 9, 10])
```

```
res = tf.multiply(x1, x2)
res
```

```
#En TF hay que abrir una sesión, para ejecutar el resultado y luego cerrarla. Es para que internamente pueda trabajar más rápido.
sess = tf.Session()
print(sess.run(res))
sess.close()
```

```
#También se puede abrir de esta forma, pero siempre se necesita estar dentro de una sesión para mostrar resultados
with tf.Session() as sess:
  output = sess.run(res)
  print(output)
```

```
#Para trabajar mejor con imágenes (son más pesadas) configuramos lo siguiente
config = tf.ConfigProto(log_device_placement = True)
config = tf.ConfigProto(allow_soft_placement = True) 
```
