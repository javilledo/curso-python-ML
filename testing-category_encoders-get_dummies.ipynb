```
import pandas as pd

url = 'https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv'
df = pd.read_csv(url)
df
```

```
import category_encoders as ce
import pandas.util.testing as tm
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

encoder = ce.BinaryEncoder(cols = ['Type 1'])
df_binary = encoder.fit_transform(df)
df_binary_test = df_binary[['Type 1_0', 'Type 1_1', 'Type 1_2', 'Type 1_3', 'Type 1_4', 'Type 1_5']]

sn.heatmap(df_binary_test.corr(), annot= True, square= True)
plt.title('Correlación con category_encoders')
```

```
plt.figure(figsize = (25,10))
df_with_dummies = pd.get_dummies(df['Type 1'], prefix = 'Type 1')
sn.heatmap(df_with_dummies.corr(), annot= True, square= True, annot_kws={"fontsize":7})
plt.title('Correlación con get_dummies')
```
