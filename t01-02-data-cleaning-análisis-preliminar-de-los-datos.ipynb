# VARIABLES DUMMY

```
data['sex']
```

```
# con la siguiente función una variable categórica la descomponemos en tantas variables como categorías tenga 
# (en este caso male y female) y esas variables serán 1 en los registros en los que tenga esa categoría
dummy_sex = pd.get_dummies(data['sex'],prefix='sex')
dummy_sex
```

```
# generalmente, se suele eliminar la variable de la que se ha generado el dummy y se adosa
# al dataset estas nuevas variables
column_name = data.columns.values.tolist()
data_wo_sex = data.drop(['sex'], axis=1)
data_wo_sex
```

```
data_concat = pd.concat([data_wo_sex,dummy_sex],axis=1)
data_concat
```

```
# Todo esto que hemos hecho de la sustitución de una variable categórica por n variables según la categoría, 
# podría hacerse a partir de una función
def create_dummies(df, var_name):
  dummy_vars = pd.get_dummies(df[var_name],prefix=var_name)
  df = data.drop([var_name], axis=1)
  df = pd.concat([df,dummy_vars],axis=1)
  return df

data_w_function = create_dummies(data,'sex')
data_w_function
```
